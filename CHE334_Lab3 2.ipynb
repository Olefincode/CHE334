```python
#Experiment1
Taveh1 = ((Tinhot1+Touthot1)/2).mean()
Tavec1 = ((Tincold1+Toutcold1)/2).mean()
print(f"For Experiment 1, evaluate heat capacity at the following average temperatures:\n Hot stream: {Taveh1} K\n Cold stream: {Tavec1} K\n")

#Experiment2
Taveh2 = ((Tinhot2+Touthot2)/2).mean()
Tavec2 = ((Tincold2+Toutcold2)/2).mean()
print(f"For Experiment 2, evaluate heat capacity at the following average temperatures:\n Hot stream: {Taveh2} K\n Cold stream: {Tavec2} K\n")
```

```python
def linearinterp(x,x1,x2,y1,y2):
  m = (y2-y1)/(x2-x1)
  return m*(x-x1)+y1
#Thermophysical values were evaluated from Welty 5 ed. Appendix I
ρh1 = linearinterp(Taveh1,313,333,992.2,983.2) #kg/m^3
ρh2 = linearinterp(Taveh2,293,313,998.2,992.2) #kg/m^3
ρc1 = linearinterp(Tavec1,293,313,998.2,992.2) #kg/m^3
ρc2 = linearinterp(Tavec2,293,313,998.2,992.2) #kg/m^3

ρ = [ρh1,ρh2,ρc1,ρc2]
V = [flow_rate * 1e-6 for flow_rate in [Vflowhot1,Vflowhot2,Vflowcold1,Vflowcold2]] #m^3/s

cph1 = linearinterp(Taveh1,313,333,4175,4181) #J/(kg*K)
cph2 = linearinterp(Taveh2,293,313,4182,4175) #J/(kg*K)
cpc1 = linearinterp(Tavec1,293,313,4182,4175) #J/(kg*K)
cpc2 = linearinterp(Tavec2,293,313,4182,4175) #J/(kg*K)

cp = [cph1,cph2,cpc1,cpc2]
ΔT = [ΔThot1,ΔThot2,ΔTcold1,ΔTcold2]

Q = []
for i in range(len(V)):
  m = ρ[i]*V[i]
  Qi = m*ΔT[i]*cp[i]
  Q.append(Qi) #W, order:


data = {
    "Experiment #": ["1","2"],
    r"Q hot [W]": [Q[0],Q[1]],
    r"Q cold [W]": [Q[2],Q[3]]
}

t = pd.DataFrame(data)
t = t.style.set_caption("Experiment 2").set_table_styles([{
    'selector': 'caption',
    'props': [
        ('color', 'black'),
        ('font-weight', 'bold')
    ]
}, {
    'selector': 'th',
    'props': [
        ('border', '1px solid black')
    ]
}, {
    'selector': 'td',
    'props': [
        ('border', '1px solid black')
    ]
}])

t
```

```python
L = 138/1000 # mm * (1 m /1000 mm) = m
Do = 6.35/1000 # mm * (1 m /1000 mm) = m
Di = 4.572/1000 # mm * (1 m /1000 mm) = m
baffl = 2/1000 # mm * (1 m /1000 mm) = m
B = 18/1000 # mm * (1 m /1000 mm) = m
Ws = 10 # mm * (1 m /1000 mm) = m
Hs = 82  # mm * (1 m /1000 mm) = m
hbw = 21 # mm * (1 m /1000 mm) = m
Nt = 2 #tubes per pass
Np = 2 #tube passes

# thermal conductivity from https://www.engineeringtoolbox.com/thermal-conductivity-plastics-d_1786.html
ko = .50 #assuming shell is made of high density polyethylene
# thermal conductivity from Welty 5 ed. Appendix H
ki = 42.9 #assuming tube is made of mild steel
```

```python
#Thermophysical values were evaluated from Welty 5 ed. Appendix I
import numpy as np
μh1 = linearinterp(Taveh1,313,333,658*10**-6,472*10**-6) #Pa*s
μh2 = linearinterp(Taveh2,293,313,993*10**-6,658*10**-6) #Pa*s
μc1 = linearinterp(Tavec1,293,313,993*10**-6,658*10**-6) #Pa*s
μc2 = linearinterp(Tavec2,293,313,993*10**-6,658*10**-6) #Pa*s

μ = [μh1,μh2,μc1,μc2]

Prh1 = linearinterp(Taveh1,313,333,4.33,3.00)
Prh2 = linearinterp(Taveh2,293,313,6.96,4.33)
Prc1 = linearinterp(Tavec1,293,313,6.96,4.33)
Prc2 = linearinterp(Tavec2,293,313,6.96,4.33)
Prhot = np.array([Prh1,Prh2])
Prcold = np.array([Prc1,Prc2])

kh1 = linearinterp(Taveh1,313,333,.633,.658) #W/(m*K)
kh2 = linearinterp(Taveh2,293,313,.597,.633) #W/(m*K)
kc1 = linearinterp(Tavec1,293,313,.597,.633) #W/(m*K)
kc2 = linearinterp(Tavec2,293,313,.597,.633) #W/(m*K)
khot = np.array([kh1,kh2])
kcold = np.array([kc1,kc2])
```

```python
#tube side = hot stream, heat transfer
Vhot = np.array([V[0],V[1]]) #Experiment 1,2
A = np.pi*(Di**2)/4
vbarhot = Vhot/(A*Nt)
ρhot = np.array([ρ[0],ρ[1]]) #Experiment 1,2
μhot = np.array([μ[0],μ[1]]) #Experiment 1,2
Rehot = ρhot*vbarhot*Di/μhot

Nuhot= .023*(Rehot**.8)*(Prhot**(1/3))
hhot = ki*Nuhot/Di


data = {
    "Experiment #": ["1","2"],
    r"Velocity [m/s]": [vbarhot[0],vbarhot[1]],
    r"Re": [Rehot[0],Rehot[1]],
    r"h [W/(m^2*K)]": [hhot[0],hhot[1]]
}

t = pd.DataFrame(data)
t = t.style.set_caption("Tube-side").set_table_styles([{
    'selector': 'caption',
    'props': [
        ('color', 'black'),
        ('font-weight', 'bold')
    ]
}, {
    'selector': 'th',
    'props': [
        ('border', '1px solid black')
    ]
}, {
    'selector': 'td',
    'props': [
        ('border', '1px solid black')
    ]
}, {
    'selector': 'td',
    'props': [
        ('border', '1px solid black')
    ]
}])

t
```

```python
#shell side = cold stream, heat transfer
Ac = B*(Ws-Do)
Ab = (hbw*Ws)-(np.pi*(Do**2)/4)

Vcold = np.array([V[2],V[3]]) #Experiment 1,2
ρcold = np.array([ρ[2],ρ[3]]) #Experiment 1,2
mcold = Vcold*ρcold

Gc = mcold/Ac
Gb = mcold/Ab
Gavg = np.sqrt(Gc*Gb)



μcold = np.array([μ[2],μ[3]]) #Experiment 1,2
Recold = Gavg*Do/μcold

Nucold = .2*(Recold**.6)*(Prcold**(1/3))
hcold = ko*Nucold/Do


data = {
    "Experiment #": ["1","2"],
    r"Gavg [m/s]": [Gavg[0],Gavg[1]],
    r"Re": [Recold[0],Recold[1]],
    r"h [W/(m^2*K)]": [hcold[0],hcold[1]]
}

t = pd.DataFrame(data)
t = t.style.set_caption("Shell-side").set_table_styles([{
    'selector': 'caption',
    'props': [
        ('color', 'black'),
        ('font-weight', 'bold')
    ]
}, {
    'selector': 'th',
    'props': [
        ('border', '1px solid black')
    ]
}, {
    'selector': 'td',
    'props': [
        ('border', '1px solid black')
    ]
}, {
    'selector': 'td',
    'props': [
        ('border', '1px solid black')
    ]
}])

t
```

```python
Thin = np.array([Tinhot1.mean(),Tinhot2.mean()]) #K
Thout = np.array([Touthot1.mean(),Touthot2.mean()]) #K
Tcin = np.array([Tincold1.mean(),Tincold2.mean()]) #K
Tcout = np.array([Toutcold1.mean(),Toutcold2.mean()]) #K
ΔTLMTDnum = (Thin-Tcout)-(Thout-Tcin)
ΔTLMTDdenom = np.log((Thin-Tcout)/(Thout-Tcin))
ΔTLMTD = ΔTLMTDnum/ΔTLMTDdenom

Ao = (np.pi*Do*L)*Nt*Np
Ai = (np.pi*Di*L)*Nt*Np

data = {
    "Experiment #": ["1","2"],
    r"ΔTLMTD [K]": [ΔTLMTD[0],ΔTLMTD[1]],
    r"Ao [m^2]": [Ao, " "],
    r"Ai [m^2]": [Ai," "]
}

t = pd.DataFrame(data)
t = t.style.set_caption("Shell-side").set_table_styles([{
    'selector': 'caption',
    'props': [
        ('color', 'black'),
        ('font-weight', 'bold')
    ]
}, {
    'selector': 'th',
    'props': [
        ('border', '1px solid black')
    ]
}, {
    'selector': 'td',
    'props': [
        ('border', '1px solid black')
    ]
}, {
    'selector': 'td',
    'props': [
        ('border', '1px solid black')
    ]
}])

t
```

```python
Rnum = Thin - Thout
Rdenom = Tcout - Tcin
R = Rnum/Rdenom
P = Rdenom/(Thin-Tcin)
print(f"R-values to evaluate correction factor, F:\n Experiment 1: {R[0]}\n Experiment 2: {R[1]}\n")
print(f"P-values to evaluate correction factor, F:\n Experiment 1: {P[0]}\n Experiment 2: {P[1]}")
```

```python
F = 1
Qh = np.array([Q[0],Q[1]])
UAexp = Qh/(ΔTLMTD*F)
k = ki
UAtheoryinv = (1/(hcold*Ao))+(np.log(Do/Di)/(2*np.pi*k*L)) + (1/(hhot*Ai))
UAtheory = (UAtheoryinv)**-1

data = {
    "Experiment #": ["1","2"],
    r"(UA)theory [W/K]": [UAtheory[0],UAtheory[1]],
    r"(UA)exp [W/K]": [UAexp[0],UAexp[1]]
}

t = pd.DataFrame(data)
t = t.style.set_caption(" ").set_table_styles([{
    'selector': 'caption',
    'props': [
        ('color', 'black'),
        ('font-weight', 'bold')
    ]
}, {
    'selector': 'th',
    'props': [
        ('border', '1px solid black')
    ]
}, {
    'selector': 'td',
    'props': [
        ('border', '1px solid black')
    ]
}])

t
```
